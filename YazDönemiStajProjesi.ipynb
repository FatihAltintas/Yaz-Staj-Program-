```python
# Gerekli kütüphaneleri yükleyelim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri kümesini yükleme
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# İlk birkaç satırı görüntüleme
print(df.head())

# Veri temizleme
print("Eksik değerler:\n", df.isnull().sum())

# Eksik değerlerin olduğu sütunları temizleme veya doldurma
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)

print("Temizlenmiş veri kümesi:\n", df.isnull().sum())

# Temel istatistiksel özetler
print(df.describe())

# Kategorik değişkenlerin dağılımı
print("Survived:\n", df['Survived'].value_counts())
print("Pclass:\n", df['Pclass'].value_counts())
print("Sex:\n", df['Sex'].value_counts())
print("Embarked:\n", df['Embarked'].value_counts())

# Veri görselleştirme
sns.set(style="whitegrid")

# Survived değişkeninin dağılımı
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.show()

# Pclass ve Survived ilişkisinin görselleştirilmesi
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Pclass')
plt.show()

# Yaş dağılımı
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.show()

# Cinsiyet ve Survived ilişkisinin görselleştirilmesi
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.show()

# Ekstra görselleştirmeler
# Yaş ve hayatta kalma durumu arasındaki ilişki
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age vs. Survival')
plt.show()

# Bilet fiyatı ve hayatta kalma durumu arasındaki ilişki
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare vs. Survival')
plt.show()

# Pclass, Age ve Fare arasındaki ilişki
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Pclass')
plt.title('Pclass, Age and Fare')
plt.show()
```

## Sonuçlar

Kadın yolcuların hayatta kalma oranı erkeklere göre daha yüksek olup, birinci sınıf yolcuların hayatta kalma oranı da diğer sınıflara göre daha yüksektir. Ayrıca, yaş ve bilet fiyatının da hayatta kalma üzerinde etkisi olduğu görülmüştür.


```python
class Item:
    def __init__(self, item_id, name, quantity):
        self.item_id = item_id
        self.name = name
        self.quantity = quantity

    def update_quantity(self, amount):
        self.quantity += amount

    def __str__(self):
        return f"ID: {self.item_id}, Name: {self.name}, Quantity: {self.quantity}"


class Inventory:
    def __init__(self):
        self.items = {}

    def add_item(self, item):
        if item.item_id in self.items:
            self.items[item.item_id].update_quantity(item.quantity)
        else:
            self.items[item.item_id] = item

    def remove_item(self, item_id):
        if item_id in self.items:
            del self.items[item_id]
        else:
            print("Item not found")

    def update_item(self, item_id, quantity):
        if item_id in self.items:
            self.items[item_id].update_quantity(quantity)
        else:
            print("Item not found")

    def display_inventory(self):
        if not self.items:
            print("No items in inventory.")
        else:
            for item in self.items.values():
                print(item)


def main():
    inventory = Inventory()

    while True:
        print("\nInventory Management System")
        print("1. Add Item")
        print("2. Remove Item")
        print("3. Update Item Quantity")
        print("4. Display Inventory")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            item_id = input("Enter item ID: ")
            name = input("Enter item name: ")
            quantity = int(input("Enter quantity: "))
            item = Item(item_id, name, quantity)
            inventory.add_item(item)
            print(f"Item added: {item}")
        elif choice == '2':
            item_id = input("Enter item ID to remove: ")
            inventory.remove_item(item_id)
        elif choice == '3':
            item_id = input("Enter item ID to update: ")
            quantity = int(input("Enter quantity to add/subtract: "))
            inventory.update_item(item_id, quantity)
        elif choice == '4':
            inventory.display_inventory()
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")


def test_inventory():
    inv = Inventory()

    item1 = Item("001", "Laptop", 10)
    item2 = Item("002", "Mouse", 50)

    inv.add_item(item1)
    inv.add_item(item2)

    assert len(inv.items) == 2  # There should be 2 items in the inventory
    assert inv.items["001"].quantity == 10  # Laptop quantity should be 10

    inv.update_item("001", 5)  # Add 5 more laptops
    assert inv.items["001"].quantity == 15  # Laptop quantity should now be 15

    inv.remove_item("002")  # Remove Mouse
    assert "002" not in inv.items  # Mouse should be removed from the inventory

    print("All tests passed!")


if __name__ == "__main__":
    main()
    # Uncomment the following line to run tests
    # test_inventory()
```

Envanter Yönetim Sistemi Raporu
1. Giriş
Bu rapor, Python kullanarak geliştirilen bir envanter yönetim sisteminin detaylarını sunmaktadır. Sistem, kullanıcıların envanterdeki öğeleri eklemesine, güncellemesine ve kaldırmasına olanak tanır. Ayrıca, envanterdeki öğelerin mevcut miktarlarını takip eder.

2. Proje Tanımı
2.1 Amaç
Bu proje, nesne yönelimli programlama prensiplerini kullanarak bir envanter yönetim sistemi geliştirmeyi amaçlamaktadır. Sistem, envanter yönetimini kolaylaştırmak için aşağıdaki işlevleri sağlar:

Öğeleri ekleme
Öğeleri güncelleme
Öğeleri kaldırma
Envanterdeki öğeleri görüntüleme
2.2 Kullanıcı Gereksinimleri
Öğelerin benzersiz bir kimlik numarası (ID), adı ve miktarı olmalıdır.
Kullanıcılar öğe ekleyebilir, kaldırabilir ve miktarlarını güncelleyebilir.
Sistem, mevcut envanter durumunu kullanıcıya gösterebilmelidir.
3. Sistem Tasarımı
3.1 Sınıflar ve Özellikler
Item Sınıfı
Amaç: Envanterdeki her bir öğeyi temsil eder.

Özellikler:

item_id: Öğenin benzersiz kimlik numarası.
name: Öğenin adı.
quantity: Öğenin mevcut miktarı.
Metodlar:

__init__: Öğeyi başlatır.
update_quantity: Miktarı günceller.
__str__: Öğenin bilgilerini string formatında döndürür.
Inventory Sınıfı
Amaç: Envanteri yönetir ve kullanıcı etkileşimlerini sağlar.

Özellikler:

items: Öğeleri saklayan bir sözlük (item_id anahtarına göre).
Metodlar:

__init__: Envanter nesnesini başlatır.
add_item: Yeni bir öğe ekler veya mevcut bir öğenin miktarını günceller.
remove_item: Belirtilen item_id'ye sahip öğeyi kaldırır.
update_item: Belirtilen item_id'ye sahip öğenin miktarını günceller.
display_inventory: Mevcut envanter durumunu ekrana yazdırır.
4. Kullanıcı Etkileşimi
4.1 Komut Satırı Arayüzü
Kullanıcı etkileşimini sağlamak için komut satırı arayüzü kullanılmıştır. Menü seçenekleri sunarak kullanıcıların:

Öğeleri eklemesine,
Öğeleri kaldırmasına,
Öğelerin miktarını güncellemesine,
Envanter durumunu görüntülemesine olanak tanır.
6. Sonuçlar ve Gelecek Adımlar
6.1 Sonuçlar
Sistem, temel envanter yönetim işlevlerini başarılı bir şekilde yerine getirmektedir.
Öğeler eklenebilir, güncellenebilir, kaldırılabilir ve envanter durumu görüntülenebilir.
